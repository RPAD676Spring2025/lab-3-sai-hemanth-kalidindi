## Inferential Statistics

### T-Tests
Compare means across two groups using the t-test. Calculate the means, conduct the t-test, and summarize the findings and what they say about the relationship between the variables.

_Write your code below:_

```python
from scipy import stats

# Splitting data into two groups
broadband_users = data_cleaned[data_cleaned['InternetAccess'] == 'Broadband']['UserSatisfaction']
mobile_data_users = data_cleaned[data_cleaned['InternetAccess'] == 'Mobile Data']['UserSatisfaction']

# Performing an independent t-test
t_stat, p_value = stats.ttest_ind(broadband_users, mobile_data_users, equal_var=False)

print(f"T-Test Results: t-statistic = {t_stat:.3f}, p-value = {p_value:.3f}")
```

1. P-value (0.655) is much greater than 0.05

This means we fail to reject the null hypothesis (no significant difference in user satisfaction between broadband and mobile data users).
In simple terms, Internet Access Type (Broadband vs. Mobile Data) does not significantly affect user satisfaction.

2. T-statistic (-0.447) is close to 0

A small t-value suggests that the difference between the two groups is minimal.

### ANOVA
Compare means across two groups using the t-test. Calculate the means, conduct the t-test, and summarize the findings and what they say about the relationship between the variables.

_Write your code below:_

```python
# Grouping data by Education Level
education_groups = [group['UserSatisfaction'] for name, group in data_cleaned.groupby('EducationLevel')]

# Conducting ANOVA test
anova_stat, anova_p = stats.f_oneway(*education_groups)

print(f"ANOVA Results: F-statistic = {anova_stat:.3f}, p-value = {anova_p:.3f}")
```

1.P-value (0.945) is much greater than 0.05

This means we fail to reject the null hypothesis (there is no significant difference in user satisfaction across different education levels).
In simple terms, education level does not significantly affect user satisfaction.

2.F-statistic (0.125) is very small

A small F-value indicates that the variability between education groups is not significantly greater than the variability within each group.
